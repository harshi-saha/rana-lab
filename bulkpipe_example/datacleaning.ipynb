# Clean the raw count files

```python
import pandas as pd
import glob
import os
import numpy as np
```

```python
#data=pd.read_csv('featurecounted/361_S49_L001.bam_p_featurecounts.txt', sep='\t')

#get all the files names  
files = glob.glob('/counts/*.genename_count.txt')

#example of reading one file into dataframe
df1 = pd.read_csv('/home/kaifuyang/Desktop/Ranalab/Server20MRepeat/counts/361_S49.genename_count.txt', sep = '\t', header = None)
df1.rename(columns={0: "geneid", 1:"counts"}, inplace=True)
df1
```

```python
#read all the files and concatenate them into one dataframe
df = pd.DataFrame()

for idx, f in enumerate(files):
    
    df2 = pd.read_csv(f, sep = '\t', header = None)
    df2.rename(columns={0: "geneid", 1:"counts" + f[-26:-19]}, inplace=True)
    
    if (idx == 0):
        df = df2
    else:
        df = pd.merge(df, df2, on = 'geneid')

df = df.sort_index(axis=1)
df.set_index('geneid', inplace = True)
df.head()
```

```python
#Since we are only comparing two groups at a time, we need to separate 106KO samples and 25KO samples
#adjust according to circumstances

df106 = df.drop(columns=['counts377_S59', 'counts378_S60', 'counts379_S61', 'counts380_S62'])
df25 = df.drop(columns = ['counts361_S49', 'counts362_S50', 'counts363_S51', 'counts364_S52'])
```

```python
#output files as csv
os.makedirs('/expression', exist_ok=True)  

df106.to_csv('/expression/miR106NTC.csv')
df25.to_csv('/expression/NTCmiR25.csv')
```
