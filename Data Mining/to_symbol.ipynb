```python
import pandas as pd

# Read the mRNA matrix file
mRNA_df = pd.read_csv('mRNAmatrix.txt', sep='\t')
```

```python
mRNA_df.sort_values(by='id')
```

```python
name_df = pd.read_csv('fff3f42f-5e42-4653-b91b-1676f914e9fa.rna_seq.augmented_star_gene_counts.tsv', sep='\t')\
[['gene_id', 'gene_name']]
```

```python
symbolTxt = pd.merge(mRNA_df, name_df, left_on='id', right_on='gene_id')\
.set_index('gene_name').drop(columns = ['id', 'gene_id']).reset_index().rename(columns={'gene_name': 'id'})
```

```python
symbolTxt.to_csv('symbol.txt', sep='\t', index=False)
```
